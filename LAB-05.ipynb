```python
#Punto 1

class Node:
    def __init__(self, value, next=None):
        assert next is None or isinstance(next,Node), "next must be Node, not %s"%(type(next))
        self.value = value
        self.next  = next
        
    def __repr__(self):
        return str(self.value)
    
class L(object):
    def __init__ (self, first_node=None):
        assert first_node is None or isinstance(first_node,Node), "first must be Node, not %s"%(type(first_node))
        self.first_node = first_node
        
    def __getitem__(self, idx):
        k = self.first_node
        for i in range(idx):
            assert k.next is not None, "index %s out of range"%(str(idx))
            k = k.next  
        return k.value
    
    def __len__(self):
        k = self.first_node
        if k is None:
            return 0
        i=1
        while k.next is not None:
            i+=1
            k = k.next
        return i
            
    def __repr__ (self):
        if self.first_node is None:
            return "[]"
        
        s = "[ %s"%self.first_node
        k=self.first_node
        while (k.next is not None):
            s += ", %s"%k.next
            k = k.next
    
        return s+" ]"

def L1(*args,**kwargs):
    
    class L1_class(L):
        
        def __getitem__(self, idx):
            actualNode = self.first_node
            if idx>=0:
              for i in range(idx):
                assert actualNode.next is not None, "Indice %s fuera de rango"%(str(idx))
                actualNode = actualNode.next
              return actualNode.value
            else:
              actualNodo = self.first_node
              numNodes = 1
              while actualNodo.next is not None:
                actualNodo  = actualNodo.next
                numNodes += 1
              assert numNodes >= -(idx), "Indice %s fuera de rango"%(str(idx))
              actualNodo = self.first_node
              for i in range(numNodes+idx):
                actualNodo = actualNodo.next
              return actualNodo.value
            
    return L1_class(*args,**kwargs)

W1 = L1(Node(10, Node(20, Node(30))))
W2 = L1(first_node=Node(10, Node(20, Node(30))))
print(W1)
print(W2)


print(W1[0], W1[1], W1[2])
print(W1[-1], W1[-2], W1[-3])

print(W1[3])

print (W1[-4])
```

```python
#Punto 4

import copy 

class Node:
    def __init__(self, value, next=None):
        assert next is None or isinstance(next,Node), "next must be Node, not %s"%(type(next))
        self.value = value
        self.next  = next
        
    def __repr__(self):
        return str(self.value)
    
class L(object):
    def __init__ (self, first_node=None):
        assert first_node is None or isinstance(first_node,Node), "first must be Node, not %s"%(type(first_node))
        self.first_node = first_node
        
    def __getitem__(self, idx):
        k = self.first_node
        for i in range(idx):
            assert k.next is not None, "index %s out of range"%(str(idx))
            k = k.next  
        return k.value
    
    def __len__(self):
        k = self.first_node
        if k is None:
            return 0
        i=1
        while k.next is not None:
            i+=1
            k = k.next
        return i
            
    def __repr__ (self):
        if self.first_node is None:
            return "[]"
        
        s = "[ %s"%self.first_node
        k=self.first_node
        while (k.next is not None):
            s += ", %s"%k.next
            k = k.next
    
        return s+" ]"

def show_ids(M, level=0):
    k = M.first_node
    while k is not None:
        print (" "*2*level, id(k))
        if (str(k.value.__class__.__name__) == str(M.__class__.__name__)):
            
            show_ids(k.value, level+1)
        k = k.next
        

def L4(*args,**kwargs):
    
    class L4_class(L):

        def clone(self):

            def clone_node(node):
              r = copy.deepcopy(node)
              return r

            r = L4(clone_node(self.first_node))
            return r
        
    return L4_class(*args,**kwargs)

W=L4(Node(10, Node(L4(Node(14, Node(15, Node(L4(Node(16, Node(17))))))), Node(20, Node(30)))))
print("Este es W: ", W)

W.__class__.__name__

K = W.clone()
print("Este es K: ", K)
print(" ")
print("Valores id de W: ")
show_ids(W)
print(" ")
print("Valores id de K: ")
show_ids(K)
```
